**瓶颈生成树（Bottleneck Spanning Tree）**

带有权重的无向图 $G(V,E,w)$ 的瓶颈生成树，表现为：在所有生成树中，最大权重边的权重值最小。即，BST $T$ 最小化瓶颈损失 $c(T)=max_{e \in T}{w(e)}$。

1. 证明 $G$ 的每一个最小生成树（MST）都是瓶颈生成树（BST）
2. 设计一个线性时间复杂度的算法：， 对于一个图 $G(V,E,w)$ 和一个整数 $b$，判断图 $ G$ 是否存在一个瓶颈生成树，其最大权重边的权重不超过 $b$，分析算法设计思路，并基于python编程实现。
3. 设计一个线性时间复杂度的算法：对于给定的图 $G(V,E,w)$，找到其瓶颈生成树，分析算法设计思路，并基于python编程实现。

answer：

idea2：
过滤边：保留所有权重 ≤b的边，形成子图 G′。

连通性检查：检查 G′是否连通。若连通，则存在满足条件的 BST。

```python
# add your code here
# algorithm of the liear time complexity：O(V+E)
def judgement(graph_edges, V, b):
    filtered_edges = [(u, v, w) for u, v, w in graph_edges if w <= b]
    adj = [[] for _ in range(V)]
    for u, v, _ in filtered_edges:
        adj[u].append(v)
        adj[v].append(u)
    visited = [False] * V
    queue = [0]
    visited[0] = True
    while queue:
        node = queue.pop(0)
        for neighbor in adj[node]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)
    return all(visited)

graph_edges = [(0,1,2), (0,2,5), (1,2,3), (1,3,4), (2,3,1)]
V = 4
b = 3
print('输入；graph_edges=', graph_edges, 'V=', V, 'b=', b)
print('输出：', judgement(graph_edges, V, b))
```

idea3：有借鉴

要达到线性时间复杂度，可以采用以下策略：
1.找到瓶颈值 b：通过二分查找或确定性选择算法，在线性时间内找到使得边权≤b 的子图恰好连通的最小 b 值。
2.构建 BST：使用边权≤b 的边构建子图，并通过 BFS/DFS 验证连通性，若连通则该子图的任意生成树即为 BST。

```python
# add your code here
# algorithm of the liear time complexity：O (E)
class UnionFind:
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [1] * size

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        fx, fy = self.find(x), self.find(y)
        if fx == fy:
            return False
        if self.rank[fx] > self.rank[fy]:
            self.parent[fy] = fx
        else:
            self.parent[fx] = fy
            if self.rank[fx] == self.rank[fy]:
                self.rank[fy] += 1
        return True

    def is_connected(self):
        root = self.find(0)
        return all(self.find(i) == root for i in range(len(self.parent)))


def find_bottleneck_tree(graph_edges, V):
    if V <= 1:
        return [], 0
    low = min(w for u, v, w in graph_edges)
    high = max(w for u, v, w in graph_edges)
    b = high
    while low <= high:
        mid = (low + high) // 2
        uf = UnionFind(V)
        for u, v, w in graph_edges:
            if w <= mid:
                uf.union(u, v)
        if uf.is_connected():
            b = mid
            high = mid - 1  # 尝试更小的b
        else:
            low = mid + 1
    filtered_edges = [(u, v, w) for u, v, w in graph_edges if w <= b]
    uf = UnionFind(V)
    bst_edges = []
    for u, v, w in filtered_edges:
        if uf.union(u, v):
            bst_edges.append((u, v, w))
            if len(bst_edges) == V - 1:
                break
    return bst_edges, b


graph_edges = [(0, 1, 2), (0, 2, 5), (1, 2, 3), (1, 3, 4), (2, 3, 1)]
V = 4
edges, bottleneck = find_bottleneck_tree(graph_edges, V)
print('输入；graph_edges=', graph_edges, 'V=', V)
print(f"输出：BST 边: {edges}")
```

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


idea：有借鉴

算法思路：
为了找到使两个城市 s和 t之间的距离减少最大的新道路，可以通过以下步骤实现：
1.预计算最短路径：在原图中计算 s 到所有节点的最短距离 Ds，以及所有节点到 t 的最短距离 Dt（通过反向图）。
2.候选边评估：对每条候选边 (u,v)，计算其可能缩短的路径长度 Ds[u]+l′(u,v)+Dt[v]，并与原最短路径比较，确定最大改进量。

时间复杂度：
预计算阶段：两次 Dijkstra 算法，时间复杂度 O(E+Vlog⁡V)。
候选边处理：线性遍历所有候选边，时间复杂度 O(K)，其中 K是候选边的数量。
总时间复杂度：O(E+Vlog⁡V+K)。

正确性分析：
1.预计算验证：Dijkstra 算法正确计算了最短路径，反向图处理确保了节点到 t的最短距离。
2.候选边评估：通过组合 s→u、候选边 u→v、v→t 的路径，确保找到可能的最优改进。
3.边界处理：跳过不可达节点，避免无效计算。

## 问题 4：

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9



idea:
算法方案：
（1）Kadane算法：遍历数组。当当前子数组和超过最大值时，更新最大值及其对应的买入和卖出日期。
（2）特殊情况处理：如果所有子数组和为负，寻找相邻两个元素的最大和，确保买入和卖出日期不同。

时间复杂度是 O (n)

````python
# add your code
# an algorithm with O(n)
class Gupiao:
    def max(self, profits):
        n = len(profits)
        if n < 2:
            return 0, None, None
        max_sum = -float('inf')
        current_sum = 0
        start = 0
        end = 0
        temp_start = 0
        for i in range(n):
            current_sum += profits[i]
            if current_sum > max_sum and i > temp_start:
                max_sum = current_sum
                start = temp_start
                end = i
            if current_sum < 0:
                current_sum = 0
                temp_start = i + 1
        if max_sum != -float('inf'):
            return max_sum, start + 1, end + 1
        else:
            max_adjacent = -float('inf')
            buy_day = 0
            sell_day = 1
            for i in range(n - 1):
                current = profits[i] + profits[i + 1]
                if current > max_adjacent:
                    max_adjacent = current
                    buy_day = i
                    sell_day = i + 1
            return max_adjacent, buy_day + 1, sell_day + 1


gupiao = Gupiao()
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = gupiao.max(profits)
print(f"输入: {profits}")
print(f"输出: 第{buy_day}天买入, 第{sell_day}天卖出,收益最大：{profit}")
````

## 问题 5：

观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)
-----
n := r-p+1
if n=2 and A[p]>A[r] then
    swap A[p] and A[r]
else if n >= 3 then
    m = ceil(2n/3)
    DoSomething(A,p,p+m-1)
    DoSomething(A,r-m+1,r)
    DoSomething(A,p,p+m-1)  
    
---
first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. 使用python编程实现上述算法或其对应的更高效版本的算法


answer:

````python
# add your code here
# your algorithm time complexity is :如上图
import math

def DoSomething(arr, p, r):
    n = r - p + 1
    if n == 2 and arr[p-1] > arr[r-1]:
        arr[p-1], arr[r-1] = arr[r-1], arr[p-1]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        DoSomething(arr, p, p + m - 1)
        DoSomething(arr, r - m + 1, r)
        DoSomething(arr, p, p + m - 1)

arr = [5, 3, 8, 4, 2, 7, 1, 6]
print("输入:", arr) 
DoSomething(arr, 1, len(arr)) 
print("输出：排序后的数组:", arr)  
````

## 问题 6：

给定一个大小为 `n` 的数组 `nums` , 返回其中的多数元素。多数元素是指在数组中出现次数 **大于** `⌊ n/2 ⌋` 的元素。

你可以假设数组是非空的, 并且给定的数组总是存在多数元素。

1. 设计一个算法找到给定数组的多数元素, 分析算法设计思路, 计算算法时间复杂度, 使用python编程实现
2. \* 设计时间复杂度为 O(n)、空间复杂度为 O(1) 的算法解决此问题, 分析算法设计思路, 使用python编程实现

e.g.:
---
1. nums=[3,2,3], 返回3
2. nums=[2,2,1,1,1,2,2], 返回2


idea1：
使用哈希表来统计每个元素的出现次数，然后找出出现次数超过数组长度一半的元素。

````python
# add your code here
# your algorithm time complexity is :O (n)
def find(nums):
    count = {}
    for num in nums:
        count[num] = count.get(num, 0) + 1
    half = len(nums) // 2
    for num, cnt in count.items():
        if cnt > half:
            return num

nums=[3,2,3]
print("输入：",nums)
print("输出：",find(nums))
````

idea for 2\*：有借鉴

Boyer-Moore 投票算法的核心思想是：在每一轮投票过程中，从数组中找出一对不同的元素并删除，直到数组为空或只有一种元素。由于多数元素的出现次数超过数组长度的一半，因此最终剩下的元素一定是多数元素。

步骤：
1.初始化候选元素为 None，计数器为 0。
2.遍历数组中的每个元素：
     如果计数器为 0，将当前元素设为候选元素。
     如果当前元素等于候选元素，则计数器加 1；否则计数器减 1。
3.遍历后，候选元素即为多数元素。

````python
# algorithm time complexity：O(n), space complexity:O(1)
# add your code here
def find(nums):
    candidate = None
    count = 0

    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)
    return candidate

nums=[2,2,1,1,1,2,2]
print("输入：",nums)
print("输出：",find(nums))
````

## 问题 7：

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1


idea1：
(A[1]>A[2])（递减），(A[n-1]<A[n])（递增），说明数组至少存在一个局部最小值。（画图可以明确看出）。

利用二分查找思想，看中间位置元素是否是局部最小值，是的话就输出这个值。不是的话，就看中间元素与左右邻居的大小，缩小范围到一半。大于右侧值说明右侧存在下降后上升的趋势，局部最小值必在右侧，在右侧继续查找局部最小值。小于右侧值说明左侧存在下降后上升趋势，局部最小值必在左侧，在左侧继续查找局部最小值。重复搜索，直到找到局部最小值或左右指针相遇，输出。

````python
# add your code here
# your algorithm time complexity is :O(logn)
def findmin(A):
    left, right = 0, len(A) - 1
    while left < right:
        mid = left + (right - left) // 2
        if mid > 0 and mid < len(A) - 1:
            if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
                return A[mid]
        if A[mid] > A[mid + 1]:
            # 右侧存在局部最小
            left = mid + 1
        else:
            # 左侧存在局部最小
            right = mid
    return A[left]


A=[9, 3, 7, 2, 1, 4, 5 ]
print('输入：', A)
print('输出：', findmin(A))
````

idea2:
算法设计思路：
遍历数组，检查每个元素是否满足局部最小值的条件（比左右邻居都小）。数组首尾元素只检查一个邻居，需判断处理。

时间复杂度：O (n)，需要遍历整个数组一次。

````python
# add your code here
# your algorithm time complexity is :O (n)
def findmin(A):
    n = len(A)
    if n == 0:
        return []
    if n == 1:
        return [A[0]]
    local_minima = []
    if A[0] < A[1]:
        local_minima.append(A[0])
    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            local_minima.append(A[i])
    if A[n - 1] < A[n - 2]:
        local_minima.append(A[n - 1])
    return local_minima

A=[9, 3, 7, 2, 1, 4, 5 ]
print('输入：', A)
print('输出：', findmin(A))
````

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3


idea1：
选择快速排序算法。

````python
# add your code here
# your algorithm time complexity is :O (n log n)
import random


def ksmall(arr, k):
    if k == 0 or not arr:
        return []
    arr = arr.copy()
    quickselect(arr, 0, len(arr) - 1, k)
    return sorted(arr[:k])


def quickselect(arr, left, right, k):
    if left >= right:
        return
    pivot_index = random.randint(left, right)
    pivot_index = partition(arr, left, right, pivot_index)
    count = pivot_index - left + 1
    if k == count:
        return
    elif k < count:
        quickselect(arr, left, pivot_index - 1, k)
    else:
        quickselect(arr, pivot_index + 1, right, k - count)


def partition(arr, left, right, pivot_index):
    # 将基准元素移到最右边
    pivot_value = arr[pivot_index]
    arr[pivot_index], arr[right] = arr[right], arr[pivot_index]
    store_index = left
    for i in range(left, right):
        if arr[i] < pivot_value:
            arr[i], arr[store_index] = arr[store_index], arr[i]
            store_index += 1
    arr[right], arr[store_index] = arr[store_index], arr[right]
    return store_index


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print('输入：', arr, ', k=', k)
print(f"输出：前{k}个最小数据为:", ksmall(arr, k))
````

idea2：
插入所有元素到堆中，执行 k 次 Extract-min 操作。

````python
# add your code here
# your algorithm time complexity is :O (n log n)
import heapq


def ksmall(arr, k):
    heap = []
    for num in arr:
        heapq.heappush(heap, num)
    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print('输入：', arr, ', k=', k)
print(f"输出：前{k}个最小数据为:", ksmall(arr, k))
````

idea3：
堆的大小不超过k,如果不超过就入，如果超过就比较，比堆顶小时就替换掉堆顶。

````python
# add your code here
# your algorithm time complexity is :O (log k)
import heapq


def ksmall(data_stream, k):
    max_heap = []
    for num in data_stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -num)
        else:
            if num < -max_heap[0]:
                heapq.heappop(max_heap)
                heapq.heappush(max_heap, -num)
    result = [-num for num in max_heap]
    result.sort()
    return result


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print('输入：', arr, ', k=', k)
print(f"输出：前{k}个最小数据为:", ksmall(arr, k))
````

举例1：数组A=["a", "da", "bde", "ab", "bc", "abdc", "cdba"], 排序后的数组应该为：['a', 'ab', 'abdc', 'bc', 'bde', 'cdba', 'da']

['a', 'ab', 'abc', 'b', 'ba', 'c']

note：

-  两个字符之间的比较可以考虑比较他们对应的ASCII码值；
- python中可以使用`ord("a")`返回字符 “a”对应的ASCII值

idea:
有借鉴。
算法设计方案：
1.构建字典树：将所有字符串插入到字典树中，每个节点表示一个字符，路径表示字符串的字符顺序，节点标记字符串的结束位置。
2.遍历字典树：通过深度优先遍历（DFS）字典树，按字符顺序（a-z）访问每个节点的子节点，收集所有字符串，确保字典序排列。

````python
# add your code here
# your algorithm time complexity is :O(n)
class Solution:
    def sortStrings(self, A):
        root = {'children': {}, 'is_end': False}
        for s in A:
            current = root
            for char in s:
                if char not in current['children']:
                    current['children'][char] = {'children': {}, 'is_end': False}
                current = current['children'][char]
            current['is_end'] = True
        result = []
        self.dfs(root, '', result)
        return result

    def dfs(self, node, prefix, result):
        if node['is_end']:
            result.append(prefix)
        for char in sorted(node['children'].keys()):
            self.dfs(node['children'][char], prefix + char, result)


def print_sort_results():
    solution = Solution()
    test_cases = [
        {
            "input": ["a", "da", "bde", "ab", "bc", "abdc", "cdba"],
            "expected": ['a', 'ab', 'abdc', 'bc', 'bde', 'cdba', 'da']
        },
        {
            "input": ['ab', 'a', 'b', 'abc', 'ba', 'c'],
            "expected": ['a', 'ab', 'abc', 'b', 'ba', 'c']
        },
        {
            "input": ['aef', 'yzr', 'wr', 'ab', 'bhjc', 'lkabdc', 'pwcdba'],
            "expected": ['ab', 'aef', 'bhjc', 'lkabdc', 'pwcdba', 'wr', 'yzr']
        }
    ]

    for idx, test in enumerate(test_cases, 1):
        input_arr = test["input"]
        expected = test["expected"]
        result = solution.sortStrings(input_arr)
        print(f"输入: {input_arr}")
        print(f"输出: {result}")
        print('')

print_sort_results()
````
